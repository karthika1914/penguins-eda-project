# Exploratory Data Analysis on the Palmer Penguins Dataset

This project explores the Palmer Penguins dataset to understand relationships between penguin species, body mass, flipper length, and bill measurements using data visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins_filepath="/content/penguins_size.csv"
penguins=pd.read_csv(penguins_filepath)
```

```python
penguins=sns.load_dataset("penguins")
penguins.head()
```

```python
penguins.info()
penguins.describe()
```

Data Cleaning - Missing values were removed to ensure accurate analysis

```python
penguins.isnull().sum()
penguins=penguins.dropna()
```

Species Distribution👇

```python
sns.countplot(data=penguins, x="species")
plt.title("Penguin Species Count")
plt.show()
```

Insight :

Adelie penguins appear most frequently in the dataset.

Body Mass Comparison 👇

```python
sns.boxplot(data=penguins, x="species", y="body_mass_g")
plt.title("Body Mass by Species")
plt.show()
```

Insight:

Gentoo penguins tend to have the highest body mass.

Feature Relationship 👇

```python
sns.scatterplot(data=penguins,x="flipper_length_mm",y="body_mass_g",hue="species")
plt.title("Flipper Length vs Body Mass")
plt.show()
```

Insight:

Penguins with longer flippers tend to have higher body mass

Correlation Heatmap 👇

```python
corr = penguins.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```

Insight :

Flipper length has the strongest correlation with body mass

Key Findings

1️)Gentoo penguins have the highest average body mass.

2️) Flipper length strongly correlates with body mass.

3️)Different species form clear clusters in scatterplots.

4️) Penguins with longer flippers tend to be heavier.
